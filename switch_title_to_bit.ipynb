```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def titleList(fileName : str):
    
    movieNum = int(fileName.split('_')[1].split('.')[0]) #int(movie)
    titleListGenre = []
    for k, v in genre_mapping.items():
        titleListGenre.append(1 if v & movieNum else 0)
        
    return titleListGenre

def get_image_data_list(folder_path, target_size=(400, 600)):
    # 지원하는 이미지 확장자
    image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')
    
    # 이미지 데이터 목록을 저장할 리스트
    image_data_list = []
    image_genre_list = []
    # 폴더 내의 모든 파일을 순회
    for filename in os.listdir(folder_path):
        # 파일의 경로 생성
        file_path = os.path.join(folder_path, filename)
        
        # 파일이 이미지 확장자 중 하나인지 확인
        if filename.lower().endswith(image_extensions) and os.path.isfile(file_path):
            # 이미지를 열고 지정된 크기로 리사이즈
            print(file_path, os.path.basename(file_path))
            img = Image.open(file_path)
            img_resized = img.resize(target_size)  # 크기 조정
            img_data = np.array(img_resized)  # 이미지를 NumPy 배열로 변환
            image_data_list.append(img_data)
            image_genre_list.append(titleList(os.path.basename(file_path)))

    return (image_data_list, image_genre_list)

# 예시 폴더 경로 (여기에 여러분의 이미지 폴더 경로를 입력하세요)
folder_path = r'G:\내 드라이브\project_poster\posters3'  
images_data, images_genre = get_image_data_list(folder_path)
# plt.imshow(images_data[0])
# 가져온 이미지 데이터 리스트의 크기 출력
for index, img_data in enumerate(images_data):
    print(f"Image {index + 1} shape: {img_data.shape} ", images_genre[index])
```

```python
import os
import re

class FileConverter:
    def __init__(self, genre_to_binary):
        """장르에 대한 이진 코드 딕셔너리를 설정"""
        self.genre_to_binary = genre_to_binary

    def convert_filename(self, old_filename):
        """파일명을 이진 코드로 변환 (장르들이 있을 경우 OR 연산 수행)"""
        # 파일명에서 장르만 추출
        genres = re.findall(r'\[([^]]+)\]', old_filename)[0].split(', ')
        
        # 각 장르의 이진 코드를 OR 연산
        binary_code = self.genre_to_binary[genres[0]]
        for genre in genres[1:]:
            # 각 장르에 해당하는 이진 코드를 가져와 OR 연산 수행
            binary_code = self.or_binary_codes(binary_code, self.genre_to_binary[genre])

        # 새로운 파일명 반환
        return f"{binary_code}.jpg"

    def or_binary_codes(self, bin1, bin2):
        """두 이진 문자열을 OR 연산"""
        # 두 이진 문자열의 비트를 비교하여 OR 연산 결과 반환
        result = ''.join(['1' if b1 == '1' or b2 == '1' else '0' for b1, b2 in zip(bin1, bin2)])
        return result

    def save_new_filename(self, old_filename):
        """파일명을 변환한 후 저장"""
        new_filename = self.convert_filename(old_filename)
        
        # 파일 경로 설정 (현재 디렉토리 기준)
        old_filepath = os.path.join(os.getcwd(), old_filename)
        new_filepath = os.path.join(os.getcwd(), new_filename)
        
        # 파일명 변경 및 저장
        if os.path.exists(old_filepath):
            os.rename(old_filepath, new_filepath)
            print(f"파일이 {new_filepath}로 저장되었습니다.")
        else:
            print(f"{old_filepath} 파일을 찾을 수 없습니다.")

# 클래스 사용 예시
genre_mapping = {
    "드라마": "1000000000000000000000000000000000000000000",
    "첩보": "0100000000000000000000000000000000000000000",
    "옴니버스": "0010000000000000000000000000000000000000000",
    "뮤직": "0001000000000000000000000000000000000000000",
    "동성애": "0000100000000000000000000000000000000000000",
    "로드무비": "0000010000000000000000000000000000000000000",
    "아동": "0000001000000000000000000000000000000000000",
    "하이틴(고교)": "0000000100000000000000000000000000000000000",
    "전기": "0000000010000000000000000000000000000000000",
    "청춘영화": "0000000001000000000000000000000000000000000",
    "재난": "0000000000100000000000000000000000000000000",
    "문예": "0000000000010000000000000000000000000000000",
    "연쇄극": "0000000000001000000000000000000000000000000",
    "신파": "0000000000000100000000000000000000000000000",
    "활극": "0000000000000010000000000000000000000000000",
    "반공/분단": "0000000000000001000000000000000000000000000",
    "군사": "0000000000000000100000000000000000000000000",
    "계몽": "0000000000000000010000000000000000000000000",
    "사회물(경향)": "0000000000000000001000000000000000000000000",
    "스포츠": "0000000000000000000100000000000000000000000",
    "합작(번안물)": "0000000000000000000010000000000000000000000",
    "종교": "0000000000000000000001000000000000000000000",
    "무협": "0000000000000000000000100000000000000000000",
    "미스터리": "0000000000000000000000010000000000000000000",
    "SF": "0000000000000000000000001000000000000000000",
    "코메디": "0000000000000000000000000100000000000000000",
    "느와르": "0000000000000000000000000010000000000000000",
    "액션": "0000000000000000000000000001000000000000000",
    "범죄": "0000000000000000000000000000100000000000000",
    "어드벤처": "0000000000000000000000000000010000000000000",
    "가족": "0000000000000000000000000000001000000000000",
    "에로": "0000000000000000000000000000000100000000000",
    "멜로/로맨스": "0000000000000000000000000000000010000000000",
    "공포": "0000000000000000000000000000000001000000000",
    "뮤지컬": "0000000000000000000000000000000000100000000",
    "시대극/사극": "0000000000000000000000000000000000010000000",
}

converter = FileConverter(genre_mapping)
converter.save_new_filename("[드라마, 미스터리, 액션].jpg")
```
