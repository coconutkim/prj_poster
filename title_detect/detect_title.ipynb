```python
import cv2
import numpy as np
import easyocr

def preprocess_image(image_path):
    # 이미지 로드
    global image
    image = cv2.imread(image_path)

    # 그레이스케일로 변환
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Canny 엣지 감지
    canny_edges = cv2.Canny(gray, 100, 200)

    # Sobel 엣지 감지
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)  # X 방향
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  # Y 방향
    sobel_edges = cv2.magnitude(sobel_x, sobel_y)

    # 두 엣지 이미지를 결합
    # combined_edges = cv2.bitwise_or(canny_edges, sobel_edges.astype(np.uint8))
    # 결합 과정에서 Sobel 대신 Canny 엣지를 더 강조
    combined_edges = cv2.addWeighted(canny_edges, 0.8, sobel_edges.astype(np.uint8), 0.2, 0)

    # 이진화 (흑백으로 변환)
    _, binary_image = cv2.threshold(combined_edges, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
      # 결과 이미지 표시
    cv2.imshow("Processed Image", binary_image)  # 수정된 부분
    cv2.waitKey(0)  # 키 입력 대기
    cv2.destroyAllWindows()  # 창 닫기

    return binary_image

def extract_text(image):
    # EasyOCR를 사용하여 텍스트 추출
    reader = easyocr.Reader(['en','ko'])  # 필요한 언어 코드 추가
    results = reader.readtext(image)

    # 결과에서 텍스트 추출
    text = " ".join([result[1] for result in results])
    
    return text.strip()

def main(image_path):
    processed_image = preprocess_image(image_path)
    
    # 결과 이미지 저장 (디버깅 용도)
    cv2.imwrite('./preprocessed/processed_edges.png', processed_image)

    title = extract_text(processed_image)
    
    print("Extracted Title:", title)

# 이미지 파일 경로 설정
image_path = 'img.jpg'  # 여기에 포스터 이미지 경로를 입력하세요
preprocess_image(image_path)
extract_text(image)
main(image_path)
```

```python
import cv2
import numpy as np

# 이미지 읽기
image = cv2.imread('shark.jpg')

# 이미지 크기 조정 (2.5배 확대)
width = int(image.shape[1] * 1)
height = int(image.shape[0] * 1)
resized_image = cv2.resize(image, (width, height))

# 그레이스케일 변환
gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# 가우시안 블러
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 소벨 엣지 검출
sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=1)  # x방향 엣지 검출
sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=1)  # y방향 엣지 검출
sobel_combined = cv2.sqrt(cv2.addWeighted(sobel_x**2, 0.5, sobel_y**2, 0.5, 0))

# 소벨 결과를 8비트로 변환
sobel_combined = cv2.convertScaleAbs(sobel_combined)

# 캐니 엣지 검출
canny_edges = cv2.Canny(gray, threshold1=100, threshold2=300)
# threshold1: 30 ~ 100
# threshold2: 100 ~ 300
# 그레이스케일 이미지에서 픽셀 값은 0부터 255까지 분포하므로,
# **threshold1**과 **threshold2**의 값은 일반적으로 0에서 255 사이로 설정됩니다.

# 소벨과 캐니 엣지 결합
combined_edges = cv2.bitwise_or(sobel_combined, canny_edges)

cv2.imwrite("./title_detect_img/combined_edges.jpg", combined_edges)

# 결과를 보여줌
cv2.imshow("Combined Edges", combined_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
