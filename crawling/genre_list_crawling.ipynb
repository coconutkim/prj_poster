```python
# 이진수 인코딩 생성
def generate_single_bit_encoding(label_texts):
    encoding = {}
    for index, genre in enumerate(label_texts):
        # 각 장르는 index 위치에 1을 설정하고 나머지는 0으로 설정
        binary_representation = ['0'] * 43  # 8비트 초기화
        binary_representation[index] = '1'  # 해당 인덱스에 1 설정
        encoding[genre] = ''.join(binary_representation)  # 리스트를 문자열로 변환
    return encoding

# 이진수 인코딩 출력
single_bit_encoding = generate_single_bit_encoding(label_texts)
for genre, binary in single_bit_encoding.items():
    print(f"{genre}: {binary}")
```

```python
genre_binary = pd.DataFrame({
        'genre':[genre for genre, binary in single_bit_encoding.items()],
        'label':[binary for genre, binary in single_bit_encoding.items()]
    })
genre_binary
```

```python
# genre와 label을 '장르=라벨' 형태로 출력
for index, row in genre_binary.iterrows():
    print(f"{row['genre']}={row['label']}")
```

```python
# 장르 이진수 표현
genre_binary = {
    '드라마': '1000000000000000000000000000000000000000000',
    '첩보': '0100000000000000000000000000000000000000000',
    '옴니버스': '0010000000000000000000000000000000000000000',
    '뮤직': '0001000000000000000000000000000000000000000',
    '동성애': '0000100000000000000000000000000000000000000'
}

# 장르를 결합하는 함수
def combine_genres(genre1, genre2):
    # 두 개의 이진수 문자열을 비트별로 OR 연산하여 결합
    combined_binary = ''
    for b1, b2 in zip(genre_binary[genre1], genre_binary[genre2]):
        combined_binary += str(int(b1) | int(b2))  # OR 연산
    return combined_binary

# 예시로 드라마와 첩보 결합
combined_genre = combine_genres('드라마', '첩보')
print(f"복합 장르 (드라마 + 첩보): {combined_genre}")
```
